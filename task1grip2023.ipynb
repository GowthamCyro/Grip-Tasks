```python
import pandas as pd #datamanipulation library
import matplotlib.pyplot as plt #plotting library
import numpy as np #powerful computation library
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split  
from sklearn.metrics import r2_score
import warnings
warnings.filterwarnings("ignore") 
```

```python
data = pd.read_csv('/kaggle/input/tsf-datasets/student_scores.csv')
```

```python
data.shape
```

```python
data.info()
```

```python
data.head()
```

```python
data.describe()
```

Visualizing the entire data

```python
data.plot(x='Hours', y='Scores', style='o')
plt.title('Hours vs Percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Scored')
plt.show()
```

Model Building

```python
X = data.iloc[:, :-1].values  
y = data.iloc[:, 1].values
```

Train_test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
print("X_train shape",X_train.shape, "\ny_train shape", y_train.shape, "\nX_test shape",X_test.shape, "\ny_test shape",y_test.shape)
```

Using Linear Regression

```python
reg = LinearRegression()  
reg.fit(X_train, y_train) 
```

```python
print("Intercept : ",reg.intercept_, '\nCoefficient',reg.coef_)
```

```python
y_pred = reg.predict(X_test)
```

Visualizing the training results 

```python
plt.scatter(X_train, y_train, color='green')
plt.plot(X_train, reg.predict(X_train), color='red')
plt.title('Hours vs Percentage (Training set)')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.grid()
plt.show()
```

```python
y_test_pred = reg.predict(X_test)
y_test_pred = pd.DataFrame(y_test_pred)
```

```python
plt.scatter(X_test, y_test_pred, color='blue')
plt.plot(X_train, reg.predict(X_train), color='green')
plt.title('Hours vs Percentage (Test set)')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.grid()
plt.show()
```

```python
print("Train : ",reg.score(X_train,y_train)*100)
print("Test : ",reg.score(X_test,y_test)*100)
```

Final Prediction

```python
hours = np.array([[9.25]])
predicted_score = reg.predict(hours)
print("Predicted Score = {}".format(predicted_score[0]))
```
